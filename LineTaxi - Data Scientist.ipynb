## 請詳述若今日我們要去開發一個陌生地區的計程車業務量，如何透過數據去了解當地潛在用戶輪廓，與如何估計當地計程車業務量需求?

### 思考架構

1. 找出目標數據，例如: 當地計程車發照數量，計程車平均薪資，空車率，上下車資訊，等等
2. 抓資料，例如: 發放問卷，當地政府的開放資料庫，計程車公會資料，叫車平台後端數據
3. 對無法抓到的資料進行假設
4. 根據資料及假設預估模型
5. 得到問題解答，並導出有價值的商業資訊

### 1. 找出目標數據

#### 用戶輪廓
- 根據計程車搭乘時機與動機區分
1. 便利性 : 目的地不好停車，行李很多，行動不便
2. 時間性 : 趕時間，半夜無其他交通工具，喝酒，下雨
3. 商業性 : 金錢充裕，商用通勤，

討論: 其他區分用戶方式，例如用戶族群 等等。

#### 用戶輪廓資料收集

1. 問卷
2. 政府資料

例如:  
1. 便利性 : 每月醫院附近，機場，飯店以及不易停車地區乘車招呼站與路欄載客次數 / 一台計程車每月平均載客次數 的百分比
2. 時間性 : 每月半夜及下大雨時載客次數 / 一台計程車每月平均載客次數 的百分比
3. 商業性 : 每月商業區附近乘車招呼站與路欄載客次數 / 一台計程車每月平均載客次數 的百分比

#### 業務量需求

1. 現有業務量統計 
2. 潛在需求

#### 業務量資料收集

1. 政府資料，計程車公會資料

例如:  
1. 現有業務量統計 : 計程車發照數量 * 計程車從業人員平均所得
2. 潛在需求 : 考量大眾運輸方便程度(計算捷運站距離，公車密集程度，等等)，平均收入，自小客車比例 等等
